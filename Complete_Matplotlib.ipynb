#Matplotlib:

*   Useful for making Plots

```python
# Importing Matplotlib Library
import matplotlib.pyplot as plt
```

```python
# Importing numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100) # will take 100 evenly spaced values between 0 and 10
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

**Plotting the Data**

```python
#sine wave
plt.plot(x,y)
plt.xlabel('angle(x)')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.xlabel('angle(x)')
plt.ylabel('cos(x)')
plt.title('Cosine Wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'r+')  #r represents red color
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'g.')  # curve with green dots
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('Parabola')
plt.show()
```

```python
plt.plot(x,y,'rx')  #official documentation contains red cross to draw curves
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('Parabola')
plt.show()
```

```python
#Plotting two curves
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine')
plt.show()
```

**Bar Plot**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.xlabel('Languages')
plt.ylabel('Number of Students')
plt.title('Bar Plot')
plt.show()
```

**Pie Chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.pie(students,labels=langs,autopct='%1.2f%%')  # 1.2 means upto 2 floating points
plt.title('Pie Chart')
plt.show()
```

**Scatter Plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

**3D Scatter Plot**

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
zdata = 20 * np.random.random(100)
xdata = np.sin(zdata)
ydata = np.cos(zdata)
ax.scatter(xdata,ydata,zdata,c=zdata,cmap='Greens')
plt.show()
```
