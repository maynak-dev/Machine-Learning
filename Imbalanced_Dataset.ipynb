#Imbalanced DataSet:
A dataset with an unequal class distribution

```python
# Importing the dependencies
import numpy as np
import pandas as pd
```

```python
# Loading the dataset to pandas dataframe
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# printing first 5 rows of the dataframe
credit_card_data.head()
```

```python
# Printing last 5 rows of the dataframe
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

0 --> Legit Transaction

1 --> Fraudulent Transaction

Here almost 99% Legit transaction as the number of legit transaction is 284315 and fraud is only 492 therefore imbalanced dataset cause we have more number of datapoints for a particular class.

```python
# separating the legit and fraudulent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**Implementing Under-Sampling Technique**

Build a sample dataset containing similar distribution of Legit and Fraudulent transaction.

No. od Fraudulent transaction --> 492

```python
#undersampling technique
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
# Concatenate the two dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)  # axis = 0 means we want to concatenate row wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

As we can see in the Class column that the legit transaction goes upward and the fraudulent transactions added to the bottom.
