**Importing the Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Imputation:**

Central Tendencies:

1.   Mean (Average Value)

2.   Median (Middle Value of the odd no. Series(ascending order) and avg. of the middle two values if even no. series)

3.   Mode (No. which has most frequency)


```python
# analyse the distribution of data in the salary column
fig, ax = plt.subplots(figsize=(10,5))
sns.distplot(dataset['salary'])
plt.show()
```

As the above plot shows it is a skew curve and we can not use mean in this. Generally, used median for this.

**Skew Curve:**
The plot generated shows a distribution of the 'salary' column. A "skew curve" or skewed distribution is a probability distribution that is not symmetrical. In the plot, the distribution is not balanced on either side of the central peak, indicating that it is skewed. Specifically, it appears to be a right-skewed (or positively skewed) distribution because the tail on the right side is longer than the tail on the left side, and the majority of the data points are clustered on the left. This means that there are some unusually high salary values that are pulling the average (mean) to the right, while the median might be a more representative measure of the central tendency for this data.

```python
# Replace the missing values with median value
dataset['salary'].fillna(dataset['salary'].median(), inplace=True)
```

**Dropping Method**

```python
# Creating another dataframe
salary_dataset = pd.read_csv('/content/Placement_Dataset.csv')
```

```python
salary_dataset.isnull().sum()
```

```python
# drop the missing values
salary_dataset = salary_dataset.dropna(how='any') # any missing values will be removed
```
